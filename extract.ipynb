```python
import random
import json
from docx import Document

```

```python
src="D:\Anantha\Academic\SEM 6\Xtra\HPE\dataset\d1.json"

with open(src, "r", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))

doc = Document()
docx_path = "QUN_ANS.docx"
```

```python
d=data['data']
print(len(d))
```

```python
qa_list=[]
para=0
if 'data' in data:
        for article in data['data']:
            if 'paragraphs' in article:  
                for idx, paragraph in enumerate(article['paragraphs']):                    
                    para+=1
                    if 'qas' in paragraph:
                        pq=paragraph['qas']
                        nums = random.sample(range(len(pq)), min(len(pq),3))
                        for ind in nums:
                            qun=pq[ind]['question']
                            ans=[a['text'] for a in pq[ind]['answers']]
                            d={"para":para,"qun":qun,"ans":ans}
                            qa_list.append(d)
                        
            
with open("Qun_Ans1.json","w")as f:
    json.dump(qa_list,f,indent=4)
```
